```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
u_cols = ['user_id', 'age', 'sex', 'occupation', 'zip_code']
users = pd.read_csv('u.user', sep='|', names=u_cols, encoding = "ISO-8859-1")

r_cols = ['user_id', 'movie_id', 'rating', 'unix_timestamp']
ratings = pd.read_csv('u.data', sep='\t', names=r_cols, encoding = "ISO-8859-1")

m_cols = ['movie_id', 'title', 'release_date', 'video_release_date', 'imdb_url']
movies = pd.read_csv('u.item', sep='|', names=m_cols, usecols=range(5), encoding = "ISO-8859-1")

#表格合并，笛卡尔积
movie_ratings = pd.merge(movies, ratings)
lens = pd.merge(movie_ratings, users)
lens
```

```python
#观察详情
lens.groupby('sex').size()
```

```python
#观察详情
lens.groupby('sex').agg({'rating': [np.size, np.mean]})
```

```python
#过滤杂项
#取投票数大于100的，不然若仅有的男女两票分别为0分与5分，则必定是分歧最大的项，需过滤
at_least100 = lens.groupby('movie_id').size()
at_least100 = at_least100[at_least100 > 100]
lens.set_index('movie_id', inplace=True)
print at_least100.head(20)
```

```python
#行选择
#MultiIndex
by_sex = lens.ix[at_least100.index].groupby(['title', 'sex'])
by_sex = by_sex.rating.mean()
by_sex
#Reshaping
```

```python
by_sex = by_sex.unstack(1) 
#取绝对值
by_sex['diff'] = abs(by_sex['F'] - by_sex['M'])
by_sex
```

```python
least_diff = by_sex.sort_values(by = 'diff')
least_diff[:5]
#least_diff.index
#least_diff.info()
```

```python
most_diff = by_sex.sort_values(by = 'diff', ascending=False)
most_diff[:5]
```
